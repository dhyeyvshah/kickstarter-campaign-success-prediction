```python
ksdata = pd.read_csv('ks-projects-201801.csv')
```

```python
ksdata = ksdata.dropna(subset=['name'])
ksdata = ksdata.drop('usd pledged', axis=1)
print(ksdata.isnull().sum())
```

```python
# Feature 1: Log transform of usd_goal_real

ksdata['goal_log'] = np.log1p(ksdata['usd_goal_real'])
print(ksdata[['usd_goal_real', 'goal_log']])
```

```python
# Feature 2: Campaign duration

ksdata['launched'] = pd.to_datetime(ksdata['launched'])
ksdata['deadline'] = pd.to_datetime(ksdata['deadline'])

ksdata['duration_days'] = (ksdata['deadline'] - ksdata['launched']).dt.days
print(ksdata[['launched', 'deadline', 'duration_days']])
```

```python
# Feature 3: One-hot encode main_category

main_category_dummies = pd.get_dummies(ksdata['main_category'], prefix='main_category', drop_first=False)

# Concatenate the dummies back to the original DataFrame
ksdata = pd.concat([ksdata, main_category_dummies], axis=1)

print(ksdata.columns)
```

```python
# Feature 4: Launch month

ksdata['launch_month'] = pd.to_datetime(ksdata['launched']).dt.month
print(ksdata[['launched', 'launch_month']])
```

```python
# Feature 5: One-hot encode country or region encoding

ksdata = pd.get_dummies(ksdata, columns=['country'])
print(ksdata.columns)
```

```python
#Feature 6: One-hot encode currency

ksdata = pd.get_dummies(ksdata, columns=['currency'])
print(ksdata.columns)
```

```python
# Feature 7: Launch day of the week

ksdata['launch_dayofweek'] = pd.to_datetime(ksdata['launched']).dt.dayofweek
print(ksdata[['launched', 'launch_dayofweek']])
```

```python
# Feature 8: Word count in project name

ksdata['name_word_count'] = ksdata['name'].apply(lambda x: len(x))
print(ksdata['name_word_count'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(data=ksdata, x='state')
plt.title('Distribution of Project Success vs. Failure')
plt.xlabel('Project State')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ksdata[ksdata['state'] == 'successful']['usd_goal_real'], bins=50, color='green', label='Successful', log_scale=True)
sns.histplot(ksdata[ksdata['state'] != 'successful']['usd_goal_real'], bins=50, color='red', label='Failed', log_scale=True)
plt.title('Distribution of Goals by Project Outcome (Log Scale)')
plt.xlabel('Goal Amount (USD)')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
filtered_data = ksdata[ksdata['state'].isin(['successful', 'failed'])]

category_counts = filtered_data.groupby(['main_category', 'state']).size().unstack()

plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar', stacked=True, color=['red', 'green'], width=0.8)
plt.title('Counts of Success and Failure by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(['Failed', 'Successful'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_goal_by_category = ksdata.groupby('main_category')['usd_goal_real'].mean().sort_values()
sns.barplot(x=avg_goal_by_category, y=avg_goal_by_category.index, palette='viridis')
plt.title('Average Goal by Main Category')
plt.xlabel('Average Goal (USD)')
plt.ylabel('Main Category')
plt.show()
```

```python
ksdata['launch_month'] = pd.to_datetime(ksdata['launched']).dt.month
success_rate_by_month = ksdata.groupby('launch_month')['state'].apply(lambda x: (x == 'successful').mean())

plt.figure(figsize=(10, 5))
sns.lineplot(x=success_rate_by_month.index, y=success_rate_by_month.values, marker='o')
plt.title('Success Rate by Launch Month')
plt.xlabel('Launch Month')
plt.ylabel('Success Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(ksdata[ksdata['state'] == 'successful']['duration_days'], label='Successful', fill=True, color='green')
sns.kdeplot(ksdata[ksdata['state'] == 'failed']['duration_days'], label='Failed', fill=True, color='red')
plt.title('Distribution of Campaign Duration by Project Outcome')
plt.xlabel('Campaign Duration (Days)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
target = ksdata['state'].apply(lambda x: 1 if x == 'successful' else 0)
features = ksdata[[
    'goal_log', 
    'duration_days', 
    'launch_month', 
    'launch_dayofweek', 
    'name_word_count'] +
    [col for col in ksdata.columns if col.startswith('main_category_')] + 
    [col for col in ksdata.columns if col.startswith('country')] +
    [col for col in ksdata.columns if col.startswith('currency')]]

X = features
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = RandomForestClassifier(class_weight={0: 1, 1: sum(y_train == 0)/sum(y_train == 1)}, random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

grid_search = GridSearchCV(
    RandomForestClassifier(class_weight={0: 1, 1: sum(y_train == 0)/sum(y_train == 1)}, random_state=42),
    param_grid,
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

print('Best Parameters:', grid_search.best_params_)

best_clf = grid_search.best_estimator_

y_pred_best = best_clf.predict(X_test)

print('Tuned Accuracy:', accuracy_score(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))
```

```python
feature_importances = best_clf.feature_importances_

features_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(features_df.head(10))

plt.figure(figsize=(10, 6))
plt.barh(features_df['Feature'], features_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.show()
```
